```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
# PROBLEM 1

# f(x) = a|x-b| + c

#------------------------------------------------------ PROBLEM 1 ------ CASE 1 -----------------------------------------#

# case 1: a1>0, b1>0, c1>0    
# a1 = 1, b1 = 1, c1 = 1

def mod1(x):
    if(x>=1):
        return 1*(x-1)+1
    else:
        return -1*(x-1)+1

def derivv_mod1(x):
    if(x>=1):
        return 1
    else:
        return -1


# initial step size
step_size = 0.1

# inital point
x = 0

Y = []
Err = []

# no. of iterations
epoch = 100

for i in range(epoch):
    x = x - step_size*derivv_mod1(x)
    # diminishing the step size by a factor of 0.9 ever iteration
    step_size = step_size*0.9 
    Y.append(mod1(x))
    Err.append(abs(x-(1)))
    
Xaxis = np.arange(100)

Err = np.array(Err)
plt.title("plot of error |x-x*| vs. iteration t for problem 1.1 where algorithm converges")
plt.xlabel("no. of iterations t")
plt.ylabel("Error")
plt.plot(Xaxis, Err)
plt.show()

Y = np.array(Y)
plt.title("plot of $f(x_{t})$ vs. iteration t for problem 1.1 where algorithm converges")
plt.xlabel("no. of iterations t")
plt.ylabel("$f(x_{t})$")
plt.plot(Xaxis, Y)
plt.show()   



#------------------------------------------------------ PROBLEM 1 ------- CASE 2 -----------------------------------------#

# case 1: a2>0, b2<0, c2>0
# a2 = 1, b2 = -1, c2 = 1

def mod2(x):
    if(x>=-1):
        return 1*(x+1)+1
    else:
        return -1*(x+1)+1

def derivv_mod2(x):
    if(x>=-1):
        return 1
    else:
        return -1

# initial step size
step_size = 0.1

# inital point
x = 0

Y = []
Err = []

# no. of iterations
epoch = 100

for i in range(epoch):
    x = x - step_size*derivv_mod2(x)
    # diminishing the step size by a factor of 0.9 every iteration
    step_size = step_size*0.9
    Y.append(mod2(x))
    Err.append(abs(x-(-1)))
    
Xaxis = np.arange(100)

Err = np.array(Err)
plt.title("plot of |x-x*| vs. iteration t for problem 1.2 where algorithm converges")
plt.xlabel("no. of iterations t")
plt.ylabel("Error")
plt.plot(Xaxis, Err)
plt.show()

Y = np.array(Y)
plt.title("plot of $f(x_{t})$ vs. iteration t for problem 1.2 where algorithm converges")
plt.xlabel("no. of iterations t")
plt.ylabel("$f(x_{t})$")
plt.plot(Xaxis, Y)
plt.show()



#------------------------------------------------------ PROBLEM 1 -------- CASE 3 -----------------------------------------#

# case 1: a3>0, b3>0, c3<0
# a3 = 1, b3 = 1, c3 = -1

def mod3(x):
    if(x>=1):
        return 1*(x-1)-1
    else:
        return -1*(x-1)-1

def derivv_mod3(x):
    if(x>=1):
        return 1
    else:
        return -1

# initial step size
step_size = 0.1

# inital point
x = 0

Y = []
Err = []

# no. of iterations
epoch = 100

for i in range(epoch):
    x = x - step_size*derivv_mod3(x)
    # diminishing the step size by a factor of 0.9 every iteration
    step_size = step_size*0.9
    Y.append(mod3(x))
    Err.append(abs(x-(1)))
    
Xaxis = np.arange(100)

Err = np.array(Err)
plt.title("plot of |x-x*| vs. iteration t for problem 1.3 where algorithm converges")
plt.xlabel("no. of iterations t")
plt.ylabel("Error")
plt.plot(Xaxis, Err)
plt.show()

Y = np.array(Y)
plt.title("plot of $f(x_{t})$ vs. iteration t for problem 1.3 where algorithm converges")
plt.xlabel("no. of iterations t")
plt.ylabel("$f(x_{t})$")
plt.plot(Xaxis, Y)
plt.show()
```

```python
# PROBLEM 3

# f(x) = x**4 + 0.5x**3 - 4x**2

def func3(x):
    return x**4 + 0.5*(x**3) - 4*(x**2)

def derivv_func3(x):
    return 4*(x**3) + 1.5*(x**2) - 8*x

# step size remains constant
step_size = 0.1



#---------------------------------------------- PROBLEM 3 CASE 1 -----------------------#

# momentum coefficient such that the convergence happens at local minima
momentum_coeff = 0.5

# no. of iterations
epoch = 100

x_t1 = 3
x_t2 = 2

Y = []
Err = []

# x value of local minima
local_min = 1.2391

for i in range(epoch):
    temp = x_t2 + momentum_coeff*(x_t2-x_t1) - step_size*derivv_func3(x_t2)
    x_t1 = x_t2
    x_t2 = temp
    Y.append(func3(x_t2))
    Err.append(abs(x_t2-local_min))
    
Xaxis = np.arange(100)

Err = np.array(Err)
plt.title("plot of |x-x*| vs. t for problem 3.1 where algorithm converges on the local minima")
plt.xlabel("t")
plt.ylabel("Error")
plt.plot(Xaxis, Err)
plt.show()

Y = np.array(Y)
plt.title("plot of $f(x_{t})$ vs. t for problem 3.1 where algorithm converges on the local minima")
plt.xlabel("t")
plt.ylabel("$f(x_{t})$")
plt.plot(Xaxis, Y)
plt.show()



#-------------------------------------------- PROBLEM 3 CASE 2 -----------------------------#

# momentum coefficiet such that algorithm escapes local minima and converges in global minima 
momentum_coeff = 0.1

# no. of iterations
epoch = 100

x_t1 = 3
x_t2 = 2

Y = []
Err = []

# x value of global minima
global_min = -1.6141

for i in range(epoch):
    temp = x_t2 + momentum_coeff*(x_t2-x_t1) - step_size*derivv_func3(x_t2)
    x_t1 = x_t2
    x_t2 = temp
    Y.append(func3(x_t2))
    Err.append(abs(x_t2-global_min))

Xaxis = np.arange(100)

Err = np.array(Err)
plt.title("plot of |x-x*| vs. iteration t for problem 3.2 where algorithm escapes local minima and converges on global minima")
plt.xlabel("no. of iterations t")
plt.ylabel("Error")
plt.plot(Xaxis, Err)
plt.show()

Y = np.array(Y)
plt.title("plot of $f(x_{t})$ vs. iteration t for problem 3.2 where algorithm escapes local minima and converges on global minima")
plt.xlabel("no. of iterations t")
plt.ylabel("$f(x_{t})$")
plt.plot(Xaxis, Y)
plt.show()



#------------------------------------------ PROBLEM 3 CASE 3 ----------------------------#

# momentum coefficient such that algorithm diverges
momentum_coeff = 0.6

# no. of iterations
# no. of iterations are only 10 since teh algorithm diverges very fast
epoch = 5

x_t1 = 3
x_t2 = 2

Y = []
Err = []

# x value of global minima
global_min = -1.6141

for i in range(epoch):
    temp = x_t2 + momentum_coeff*(x_t2-x_t1) - step_size*derivv_func3(x_t2)
    x_t1 = x_t2
    x_t2 = temp
    Y.append(func3(x_t2))
    Err.append(abs(x_t2-global_min))

Xaxis = np.arange(5)

Err = np.array(Err)
plt.title("plot of |x-x*| vs. iteration t for problem 3.3 where algorithm escapes global minima too and diverges")
plt.xlabel("no. of iterations t")
plt.ylabel("Error")
plt.plot(Xaxis, Err)
plt.show()

Y = np.array(Y)
plt.title("plot of $f(x_{t})$ vs. iteration t for problem 3.3 where algorithm escapes global minima too and diverges")
plt.xlabel("no. of iterations t")
plt.ylabel("$f(x_{t})$")
plt.plot(Xaxis, Y)
plt.show()
```

```python
# problem 4
# f(x) = x**4 + 0.5x**3 - 4x**2

def func4(x):
    return x**4 + 0.5*(x**3) - 4*(x**2)

def derivv_func4(x):
    return 4*(x**3) + 1.5*(x**2) - 8*x

def second_derivv_func4(x):
    return 12*(x**2) + 3*(x) - 8

#-------------------------------------------- PROBLEM 4 CASE 1 --------------------------------#

# inital point such that it converges to local minima
x = 10

# no. of iterations
epoch = 10

# x value of local minima 
local_min = 1.2391

Y = []
Err = []

for i in range(epoch):
    x = x - derivv_func4(x)/second_derivv_func4(x)
    Y.append(func4(x))
    Err.append(abs(x-local_min))
    
Xaxis = np.arange(10)

Err = np.array(Err)
plt.title("plot of |x-x*| vs. t for problem 4.1 where algorithm converges on the local minima")
plt.xlabel("t")
plt.ylabel("Error")
plt.plot(Xaxis, Err)
plt.show()

Y = np.array(Y)
plt.title("plot of $f(x_{t})$ vs. t for problem 4.1 where algorithm converges on the local minima")
plt.xlabel("t")
plt.ylabel("$f(x_{t})$")
plt.plot(Xaxis, Y)
plt.show()  

#------------------------------------------- PROBLEM 4 CASE 2 --------------------------------#

# inital point such that it converges to global minima
x = -10

# x value of global minima
global_min = -1.6141

Y = []
Err = []

# no. of iterations
epoch = 10

for i in range(epoch):
    x = x - derivv_func4(x)/second_derivv_func4(x)
    Y.append(func4(x))
    Err.append(abs(x-global_min))
    
Xaxis = np.arange(10)

Err = np.array(Err)
plt.title("plot of |x-x*| vs. iteration for problem 4.2 where algorithm converges on the global minima")
plt.xlabel("no. of iterations t")
plt.ylabel("Error")
plt.plot(Xaxis, Err)
plt.show()

Y = np.array(Y)
plt.title("plot of $f(x_{t})$ vs. iteration for problem 4.2 where algorithm converges on the global minima")
plt.xlabel("no. of iterations t")
plt.ylabel("$f(x_{t})$")
plt.plot(Xaxis, Y)
plt.show()
```
